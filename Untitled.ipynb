```python
import csv
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
```

```python



def calc_average_price(row):
    if pd.isna(row["Average price"]):
        return float((row["Max price"] + row["Min price"])/2)
    else:
        return row["Average price"]

dataFrame = pd.read_csv('train.csv')

dataFrame["Average price"] = dataFrame.apply(calc_average_price, axis=1)

categories = pd.DataFrame(dataFrame['Category'].str.split('/', expand=True).values.tolist(), index=dataFrame.index)

dataFrame[['category','sub_category','sub_sub_category', 'sub_sub_sub_category']] = categories[[1, 2, 3, 4]]

category = dataFrame["category"].unique()
sub_category = dataFrame["sub_category"].unique()
sub_sub_category = dataFrame["sub_sub_category"].unique()
sub_sub_sub_category = dataFrame["sub_sub_sub_category"].unique()
brands = dataFrame["Brand"].unique()
sellers = dataFrame["Seller"].unique()


def change_brand_name_to_ordinal(row):
    return next(iter(np.where(brands == row["Brand"])[0].tolist()), 0)

def change_seller_name_to_ordinal(row):
    return next(iter(np.where(sellers == row["Seller"])[0].tolist()), 0)

def change_category_name_to_ordinal(row):
    return next(iter(np.where(category == row["category"])[0].tolist()), 0)

def change_sub_category_name_to_ordinal(row):
    return next(iter(np.where(sub_category == row["sub_category"])[0].tolist()), 0)

def change_sub_sub_category_name_to_ordinal(row):
    return next(iter(np.where(sub_sub_category == row["sub_sub_category"])[0].tolist()), 0)

def change_sub_sub_sub_category_name_to_ordinal(row):
    return next(iter(np.where(sub_sub_sub_category == row["sub_sub_sub_category"])[0].tolist()), 0)

def replace_rating_comma_to_dot(row):
    return row["Rating"].replace(',','.')


dataFrame["Brand"] = dataFrame.apply(change_brand_name_to_ordinal, axis=1)
dataFrame["Seller"] = dataFrame.apply(change_seller_name_to_ordinal, axis=1)
dataFrame["Rating"] = dataFrame.apply(replace_rating_comma_to_dot, axis=1)

dataFrame["category"] = dataFrame.apply(change_category_name_to_ordinal, axis=1)
dataFrame["sub_category"] = dataFrame.apply(change_sub_category_name_to_ordinal, axis=1)
dataFrame["sub_sub_category"] = dataFrame.apply(change_sub_sub_category_name_to_ordinal, axis=1)
dataFrame["sub_sub_sub_category"] = dataFrame.apply(change_sub_sub_sub_category_name_to_ordinal, axis=1)


dataFrame = dataFrame.drop(['Color', 'full_category', 'Name', 'Category'], axis=1)

# dataFrame["Brand"][0].type

# categories
dataFrame
```

```python
import array

n_input, n1_hidden, n2_hidden, n_out, batch_size, learning_rate = 17, 34, 51, 1, 100, 0.0000001


model = nn.Sequential(
    nn.Linear(n_input, n1_hidden),
    nn.ReLU(),
    nn.Linear(n1_hidden, n2_hidden),
    nn.ReLU(),
    nn.Linear(n2_hidden, n1_hidden),
    nn.ReLU(),
    nn.Linear(n1_hidden, n_out)
)

model.double()

loss_function = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

dataFrame = dataFrame.astype(np.float32)

x = dataFrame.drop(["Sales"], axis = 1).to_numpy()[1:]
y = dataFrame["Sales"].to_numpy()[1:]
```

```python
losses = []
for epoch in range(1000):
    print(epoch)
    for i in range(1, dataFrame.shape[0]):
        data_x = torch.tensor(np.array(x[i-1].tolist()))
        data_y = torch.tensor(np.array(y[i-1].tolist()))
        pred_y = model(data_x)
        loss = loss_function(pred_y, data_y)
        losses.append(loss.item())

        model.zero_grad()
        loss.backward()

        optimizer.step()
```
